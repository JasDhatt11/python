```python
names = [' Jas', 'Karan ', 'tejinder', 'jashan', 'Param', '', 'Harry']
cleaned_names = []

for name in names:
  cleaned = name.strip()
  if cleaned == '':
     cleaned_names.append("Unknown")
  else:cleaned_names.append(cleaned.title())
print("Cleaned Names:")
print(cleaned_names)

```

```python
emails = ["  JAS@GMAIL.COM ", "", "user@yahoo.com", "  "]
cleaned_emails = []

for email in emails:
    cleaned = email.strip()
    if cleaned == '':
        cleaned_emails.append('unknown')
    else:
        cleaned_emails.append(cleaned.title())

print('Cleaned Emails')
print(cleaned_emails)
```

```python
phones = [" 9876543210", "12345", "", "9123456789 "]
cleaned_phones = []

for phone in phones:
    cleaned = phone.strip()
    if cleaned == '': 
        cleaned_phones.append('unknown')
    else:
        cleaned_phones.append(cleaned.title())

print('Cleaned Phones')
print(cleaned_phones)
```

```python
phones = [" 9876543210", "12345", "", "9123456789 "]
cleaned_phones = []

for phone in phones:
    cleaned = phone.strip()
    if len(cleaned) == 10 and cleaned.isdigit(): 
        
        formatted = cleaned[0:3] + "-" + cleaned[3:6] + "-" + cleaned[6:]
        cleaned_phones.append(formatted)
    else:
          cleaned_phones.append("Invalid")
        
print(cleaned_phones)
```
